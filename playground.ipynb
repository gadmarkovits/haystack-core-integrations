```python
from haystack_integrations.components.embedders.opea import OPEADocumentEmbedder, OPEATextEmbedder
from haystack import Document, Pipeline

doc = Document(content="I love pizza!")

doc_embedder = OPEADocumentEmbedder(api_url="http://localhost:8090")
doc_embedder.warm_up()

result = doc_embedder.run([doc])
print(result["documents"][0].embedding)
```

```python
text_embedder = OPEATextEmbedder()
text_pipeline = Pipeline()
text_pipeline.add_component("embedder", text_embedder)
text_pipeline.run({"text": "I like pizza!"})
```
